```python
import numpy as np
import pandas as pd
import powerlaw 
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
dataset= pd.read_excel(r'Humboldt-FAO.xlsx')
dataset
```

```python
# Normalization
def stand_sca(data):
    new_data=(data-data.mean())/data.std()
    return new_data
```

```python
# Use the 'ewstools' package
import ewstools
```

```python
# read the data from dataset, EWS are not computed beyond the transition time
ts = ewstools.TimeSeries(data=dataset['A.landing-sum'], transition=65)
ts.detrend(method='Gaussian', bandwidth=0.1)
ts.compute_var(rolling_window=0.25)
ts.compute_auto(lag=1, rolling_window=0.25)
ts.compute_auto(lag=2, rolling_window=0.25)
ts.compute_auto(lag=3, rolling_window=0.25)
ts.compute_ktau()
# plot the results
ts.make_plotly()
```
