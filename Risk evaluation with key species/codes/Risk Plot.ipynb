```python
import matplotlib.pyplot as plt
import numpy as np

# Load risk data for each current
currents = ['Benguela', 'California', 'Humboldt', 'Kuroshio-Oyashio', 'NE-Atlantic', 'Brazil-Argentina']
risks = [0.638189908,0.099471037,1.011160919,100,18.25167095]

# Normalize the risks
max_risk = max(risks)
normalized_risks = [risk / max_risk * 100 for risk in risks]

# Sort the data by risk in descending order
sorted_data = sorted(zip(currents, normalized_risks), key=lambda x: x[1], reverse=False)
sorted_currents, sorted_normalized_risks = zip(*sorted_data)

# Generate colors from green to light green based on risk values
colors = plt.cm.Purples(np.linspace(0.3, 1, 5))

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(sorted_currents, sorted_normalized_risks, color=colors)

# Add the risk values at the end of each bar
for bar, risk in zip(bars, sorted_normalized_risks):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{risk:.2f}', 
             va='center', ha='left')

# Remove the x-axis ticks
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Add the x-axis label
plt.xlabel('Risk Value')


plt.title('Normalized Anchovy Risk', fontsize=16)
# Remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Show the plot
```
