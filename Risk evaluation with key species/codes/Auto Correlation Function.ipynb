```python
import numpy as np
import pandas as pd
```

```python
# Generate clips
def Generate(dataframe, step, cliplen):
    
    Len = len(dataframe)
    
    # number of clips
    num = (Len - cliplen) // step + 1
    
    dfs = []
    
    for i in range(0, num):
        df = dataframe.loc[i * step : i * step + cliplen - 1]
        dfs.append(df)
        
    return dfs, num
```

```python
def stand_sca(data):
    
    # Normalization with sca
    new_data=(data-data.mean())/data.std()
    return new_data
```

```python
# import data
df_result = pd.read_excel(r'Benguela-FAO.xlsx')
df_result = df_result.drop(columns=['year','region-FAO'])
df_result = stand_sca(df_result)
df_result.head()
```

```python
all_clips, number = Generate(df_result, 6, 20)
```

```python
# import the required modules
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df_result['A.landing-sum'])
# plot the auto correlation
for clip in all_clips:
    plot_acf(clip['A.landing-sum'])
```
